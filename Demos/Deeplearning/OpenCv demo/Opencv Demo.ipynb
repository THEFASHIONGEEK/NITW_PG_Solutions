# Face detection using opencv



```python
import cv2


import cv2
##C:\Users\Sampriti_Chatterjee\PycharmProjects\CNN_binary\haarcascade_frontalface_default.xml
# Load the cascade
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
# Read the input image
img = cv2.imread('download123.jpg')
# Convert into grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.1, 4)
# Draw rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
# Display the output
cv2.imshow('img', img)
cv2.waitKey()
```

# Video capturing using OpenCV



```python
import cv2,time
video=cv2.VideoCapture(0)
#create VideoCapture object to trigger the camera
# 0 is to specify that use built-in camera
if (video.isOpened() == False):
    print("Unable to read camera feed")

check , frame = video.read()
print(check)
print(frame)
time.sleep(3) # will stop the script for 3 seconds
video.release()
#This will release the camera in some milliseconds

```
